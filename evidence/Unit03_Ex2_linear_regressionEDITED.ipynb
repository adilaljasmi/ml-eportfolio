```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np

print("=== EXPERIMENT 1: ORIGINAL SMALL DATASET ===")

# Original small dataset
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

# Execute linear regression
slope, intercept, r, p, std_err = stats.linregress(x, y)

# Measure correlation
corr, _ = stats.pearsonr(x, y)
print('Pearsons correlation: %.3f' % corr)
print('R-squared: %.3f' % (r**2))
print('Slope: %.3f' % slope)
print('Intercept: %.3f' % intercept)
print('P-value: %.3f' % p)

# Create prediction function
def myfunc(x):
    return slope * x + intercept

# Generate predictions
mymodel = list(map(myfunc, x))

# Create figure with subplots
plt.figure(figsize=(15, 5))

# Plot 1: Original data
plt.subplot(1, 3, 1)
plt.scatter(x, y, color='blue', alpha=0.7)
plt.plot(x, mymodel, color='red', linewidth=2)
plt.title('Original Small Dataset (n=13)')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True, alpha=0.3)

print()

print("=== EXPERIMENT 2: LARGER DATASET WITH MORE NOISE ===")

# Generate larger dataset with more noise
np.random.seed(42)
x_large = np.random.uniform(1, 20, 100)  # 100 random points
y_large = -2 * x_large + 90 + np.random.normal(0, 15, 100)  # Add significant noise

# Linear regression on larger dataset
slope2, intercept2, r2, p2, std_err2 = stats.linregress(x_large, y_large)
corr2, _ = stats.pearsonr(x_large, y_large)

print('Pearsons correlation: %.3f' % corr2)
print('R-squared: %.3f' % (r2**2))
print('Slope: %.3f' % slope2)
print('Intercept: %.3f' % intercept2)
print('P-value: %.3f' % p2)

# Plot 2: Larger noisy dataset
plt.subplot(1, 3, 2)
plt.scatter(x_large, y_large, color='orange', alpha=0.6)
x_line = np.linspace(min(x_large), max(x_large), 100)
y_line = slope2 * x_line + intercept2
plt.plot(x_line, y_line, color='red', linewidth=2)
plt.title('Large Noisy Dataset (n=100)')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True, alpha=0.3)

print()

print("=== EXPERIMENT 3: PERFECT LINEAR RELATIONSHIP ===")

# Generate perfect linear relationship
x_perfect = np.linspace(1, 15, 50)
y_perfect = -1.5 * x_perfect + 95  # No noise

slope3, intercept3, r3, p3, std_err3 = stats.linregress(x_perfect, y_perfect)
corr3, _ = stats.pearsonr(x_perfect, y_perfect)

print('Pearsons correlation: %.3f' % corr3)
print('R-squared: %.3f' % (r3**2))
print('Slope: %.3f' % slope3)
print('Intercept: %.3f' % intercept3)
print('P-value: %.3f' % p3)

# Plot 3: Perfect linear relationship
plt.subplot(1, 3, 3)
plt.scatter(x_perfect, y_perfect, color='green', alpha=0.8)
plt.plot(x_perfect, y_perfect, color='red', linewidth=2)
plt.title('Perfect Linear Relationship')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print()
print("=== LEARNING OBSERVATIONS ===")
print(f"Original small dataset R²: {r**2:.3f}")
print(f"Large noisy dataset R²: {r2**2:.3f}")
print(f"Perfect relationship R²: {r3**2:.3f}")
print()
print("Professional/Ethical Implications:")
print("1. Small sample sizes can give misleading regression confidence")
print("2. Noise in real-world data affects prediction accuracy")
print("3. Perfect correlations rarely exist - beware of overfitting")
print("4. P-values help assess statistical significance of relationships")
print("5. R² values indicate how much variance the model explains")

# Prediction example using original function
print()
print("=== PREDICTION EXAMPLE ===")
test_value = 10
prediction = myfunc(test_value)
print(f"Prediction for x={test_value}: y={prediction:.1f}")
print("Note: Predictions outside the data range may be unreliable")
```
