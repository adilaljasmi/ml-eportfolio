```python
# calculate the Pearson's correlation between two variables
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot as plt
import seaborn as sns

from scipy.stats import pearsonr

# EXPERIMENT 1: Original parameters
print("=== EXPERIMENT 1: ORIGINAL PARAMETERS ===")
seed(1)
data1 = 20 * randn(1000) + 100
data2 = data1 + (10 * randn(1000) + 50)

covariance = cov(data1, data2)
corr, _ = pearsonr(data1, data2)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.scatter(data1, data2, alpha=0.6)
plt.title('Original: Strong Correlation')
plt.xlabel('Data1')
plt.ylabel('Data2')

print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))
print('Covariance: %.3f' % covariance[0][1])
print('Pearsons correlation: %.3f' % corr)
print()

# EXPERIMENT 2: Increased noise (weaker correlation)
print("=== EXPERIMENT 2: INCREASED NOISE ===")
seed(1)
data1_v2 = 20 * randn(1000) + 100
data2_v2 = data1_v2 + (40 * randn(1000) + 50)  # Increased noise from 10 to 40

covariance_v2 = cov(data1_v2, data2_v2)
corr_v2, _ = pearsonr(data1_v2, data2_v2)

plt.subplot(1, 3, 2)
plt.scatter(data1_v2, data2_v2, alpha=0.6, color='orange')
plt.title('Increased Noise: Weaker Correlation')
plt.xlabel('Data1')
plt.ylabel('Data2')

print('data1: mean=%.3f stdv=%.3f' % (mean(data1_v2), std(data1_v2)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2_v2), std(data2_v2)))
print('Covariance: %.3f' % covariance_v2[0][1])
print('Pearsons correlation: %.3f' % corr_v2)
print()

# EXPERIMENT 3: Smaller sample size
print("=== EXPERIMENT 3: SMALLER SAMPLE SIZE ===")
seed(1)
data1_v3 = 20 * randn(100) + 100  # Reduced from 1000 to 100 samples
data2_v3 = data1_v3 + (10 * randn(100) + 50)

covariance_v3 = cov(data1_v3, data2_v3)
corr_v3, _ = pearsonr(data1_v3, data2_v3)

plt.subplot(1, 3, 3)
plt.scatter(data1_v3, data2_v3, alpha=0.8, color='green')
plt.title('Smaller Sample: N=100')
plt.xlabel('Data1')
plt.ylabel('Data2')

plt.tight_layout()
plt.show()

print('data1: mean=%.3f stdv=%.3f' % (mean(data1_v3), std(data1_v3)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2_v3), std(data2_v3)))
print('Covariance: %.3f' % covariance_v3[0][1])
print('Pearsons correlation: %.3f' % corr_v3)
print()

# SUMMARY OF OBSERVATIONS
print("=== LEARNING OBSERVATIONS ===")
print(f"Original correlation: {corr:.3f}")
print(f"With increased noise: {corr_v2:.3f}")
print(f"With smaller sample: {corr_v3:.3f}")
print()
print("Professional/Ethical Implications:")
print("1. Sample size affects reliability of correlation estimates")
print("2. Noise in data can mask true relationships")
print("3. Correlation does not imply causation - ethical responsibility in reporting")
print("4. Small samples may lead to misleading conclusions in ML applications")
```
